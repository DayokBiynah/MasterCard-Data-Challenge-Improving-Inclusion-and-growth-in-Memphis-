```python
# heatmap_save.py
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# -------------------------------------------------
# 1. Entering south memphis data, with urban-rural and state benchmark
# -------------------------------------------------
data = {
    'Benchmark': [
        'Net Occupancy, %','Residential Real Estate Value','Acres of Park Land, %',
        'Affordable Housing, %','Internet Access, %','Travel Time to Work, %',
        'New Businesses, %','Spend Growth, %','Small Business Loans, %',
        '% Women Owned Business','Labor Market Engagement Index','Commercial Diversity, %',
        'Personal Income, %','Spending per Capita, %','People Above Poverty Base',
        'Gini Coefficient Base','Early Education Enrollment','Health Insurance Coverage'
    ],
    'South Memphis Tract': [
        -2.25,13.35,3.32,76.53,43.5,77.05,9.87,0,14.56,7.38,10.44,12.84,10.02,0,59.21,47.73,23.5,81.76
    ],
    'Urban-Rural Base': [
        0.78,25.2,4.1,72.39,87.21,56.53,5.78,0,3.98,4.09,59.88,18.26,11.51,0,87.83,41.45,26.53,93.86
    ],
    'State Base': [
        0.99,24.19,2.9,80.33,78.24,66.95,7.76,0,6.47,3.58,38.25,17.34,10.9,0,82.88,42.6,22.08,90.55
    ]
}
df = pd.DataFrame(data)

# Set the index **before** the heatmap 
df.set_index('Benchmark', inplace=True)
```

```python
# -------------------------------------------------
# 2. Heatmap
# -------------------------------------------------
plt.figure(figsize=(10, 12))
sns.set(font_scale=1.1)

ax = sns.heatmap(
    df,
    annot=True,
    fmt='.2f',
    cmap='RdYlGn',
    linewidths=0.5,
    linecolor='gray',
    cbar_kws={'label': 'Value', 'shrink': 0.8},
    vmin=df.values.min(),
    vmax=df.values.max()
)
```

```python
# -------------------------------------------------
# 3. **CRITICAL**  keep the axis ON while saving
# -------------------------------------------------
plt.tight_layout()                     # removes excess whitespace
plt.savefig(
    'south_memphis_urban_state_heatmap.png',
    dpi=300,
    bbox_inches='tight',               # crops only the outer padding
    facecolor='white'                  # guarantees a white background
)
plt.close()   # no pop-up window, no memory leak

print("Heatmap saved as 'south_memphis_heatmap.png'")
```
