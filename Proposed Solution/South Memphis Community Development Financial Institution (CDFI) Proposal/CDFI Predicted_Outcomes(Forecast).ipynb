```python
import matplotlib.pyplot as plt
import numpy as np

# Years: 2024.0 to 2028.0 (mid-year points)
years = np.array([2024.0, 2024.5, 2025.0, 2025.5, 2026.0, 2026.5, 2027.0, 2027.5, 2028.0])

# Baseline: flat at 34 (no change)
baseline = np.full(len(years), 34)

# With CDFI: starts at 34 (2024), reaches 45 by 2026, then +1.5 pts/year
with_cdfi = [34, 36.5, 39, 41.5, 45, 46.5, 48, 49.5, 51]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(years, baseline, 'r--s', label='Baseline', markersize=6)
plt.plot(years, with_cdfi, 'g-o', label='With CDFI', markersize=8, linewidth=3)

# Style
plt.title("South Memphis IGS Forecast: Mastercard Data Challenge", fontsize=14, fontweight='bold')
plt.xlabel("Year")
plt.ylabel("IGS Score")
plt.ylim(32, 52)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Labels
plt.text(2024.0, 34.3, '34', fontsize=10)
plt.text(2026.0, 45.3, '45', fontsize=10, color='green', fontweight='bold')

# Save
plt.tight_layout()
plt.savefig("igs_forecast_fixed.png", dpi=300)
plt.show()
```
